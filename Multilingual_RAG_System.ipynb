```

# CELL 1: INSTALL ALL NECESSARY DEPENDENCIES

print("⏳ Installing required Python libraries...")
!pip install langchain langchain-community langchain-google-genai transformers torch sentence-transformers faiss-cpu pydantic rank_bm25 --quiet
print("✅ All installations complete.")



# CELL 2: SET UP API KEY AND IMPORT LIBRARIES

import os
import torch
from google.colab import userdata, files

from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.docstore.document import Document
from langchain.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.cross_encoders import HuggingFaceCrossEncoder
from langchain.retrievers.document_compressors import CrossEncoderReranker
from langchain.retrievers import ContextualCompressionRetriever, EnsembleRetriever
from langchain_community.retrievers import BM25Retriever
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.prompts import PromptTemplate
from langchain.schema.runnable import RunnablePassthrough
from langchain.schema.output_parser import StrOutputParser

# Securely get the API key from the user
try:
    os.environ["GOOGLE_API_KEY"] = userdata.get('GOOGLE_API_KEY')
except userdata.SecretNotFoundError:
    api_key = input("Please enter your Google API Key: ")
    os.environ["GOOGLE_API_KEY"] = api_key
    userdata.set('GOOGLE_API_KEY', api_key)

print("✅ Google API Key set successfully.")
print("✅ All libraries imported.")



# CELL 3: UPLOAD YOUR KNOWLEDGE BASE (TEXT FILE)
# Run this cell and select your 'aparichita.txt' file.

TEXT_FILE_PATH = "aparichita.txt"

print(f"Please upload your clean text file. It should be named '{TEXT_FILE_PATH}'.")
uploaded = files.upload()

if TEXT_FILE_PATH in uploaded:
    print(f"✅ Successfully uploaded '{TEXT_FILE_PATH}'.")
else:
    print(f"❌ File '{TEXT_FILE_PATH}' not found in upload. Please rename your file and run this cell again.")



# CELL 4: DEFINE DATA PROCESSING AND RAG PIPELINE FUNCTIONS

FAISS_INDEX_PATH = "faiss_index_txt"
EMBEDDING_MODEL_NAME = 'sentence-transformers/LaBSE'
RERANKER_MODEL_NAME = 'cross-encoder/ms-marco-MiniLM-L-6-v2'
LLM_MODEL_NAME = 'gemini-1.5-flash'

# Global variable to hold chunks for BM25
text_chunks = []

def build_vector_store_from_txt():
    """Builds and saves a FAISS vector store and prepares chunks for BM25."""
    global text_chunks
    if not os.path.exists(TEXT_FILE_PATH):
        print(f"ERROR: Text file not found at '{TEXT_FILE_PATH}'. Please upload it first.")
        return False

    print("--- Starting Data Ingestion Pipeline from TXT file ---")

    with open(TEXT_FILE_PATH, 'r', encoding='utf-8') as f:
        text_content = f.read()

    if not text_content:
        print("ERROR: The text file is empty. Aborting.")
        return False

    print("INFO: Chunking documents...")
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
    docs = [Document(page_content=text_content)]
    text_chunks = text_splitter.split_documents(docs)
    print(f"INFO: Created {len(text_chunks)} text chunks.")

    print(f"INFO: Initializing embedding model: {EMBEDDING_MODEL_NAME}...")
    embeddings = HuggingFaceEmbeddings(model_name=EMBEDDING_MODEL_NAME)

    print("INFO: Building FAISS vector store... (This may take a minute)")
    vector_store = FAISS.from_documents(text_chunks, embeddings)

    print(f"INFO: Saving FAISS index to '{FAISS_INDEX_PATH}'...")
    vector_store.save_local(FAISS_INDEX_PATH)
    print("--- ✅ Knowledge Base built successfully! ---")
    return True

def get_rag_pipeline(retriever):
    """Builds the RAG chain using a provided retriever."""
    print("INFO: Loading RAG pipeline components...")

    llm = ChatGoogleGenerativeAI(model=LLM_MODEL_NAME, temperature=0.1)

    prompt_template = """
    You are a helpful assistant. Answer the user's question based ONLY on the following context.
    Your response must be in the same language as the user's question.
    If the context does not contain the answer, state that you cannot find the answer in the provided documents.

    CONTEXT:
    {context}

    QUESTION:
    {question}

    ANSWER:
    """
    prompt = PromptTemplate(template=prompt_template, input_variables=["context", "question"])

    rag_chain = (
        {"context": retriever, "question": RunnablePassthrough()}
        | prompt
        | llm
        | StrOutputParser()
    )

    print("✅ RAG Pipeline is ready.")
    return rag_chain


# CELL 5: BUILD THE KNOWLEDGE BASE FROM THE TEXT FILE
# Run this cell to process your uploaded .txt file.

build_success = build_vector_store_from_txt()


# CELL 6: DEBUGGING - ISOLATE AND TEST RETRIEVERS
# This cell tests each retriever independently to see what they are finding.

if build_success:
    print("\n--- Starting Retriever Debugging Session ---")

    # Load the FAISS vector store
    embeddings = HuggingFaceEmbeddings(model_name=EMBEDDING_MODEL_NAME)
    vector_store = FAISS.load_local(FAISS_INDEX_PATH, embeddings, allow_dangerous_deserialization=True)

    # 1. Test Semantic Search Retriever (FAISS)
    faiss_retriever = vector_store.as_retriever(search_kwargs={"k": 2})

    # 2. Test Keyword Search Retriever (BM25)
    bm25_retriever = BM25Retriever.from_documents(text_chunks)
    bm25_retriever.k = 2

    test_query = "অনুপমের ভাগ্য দেবতা কে"

    print(f"\n--- Testing FAISS (Semantic) Retriever with query: '{test_query}' ---")
    faiss_results = faiss_retriever.invoke(test_query)
    for i, doc in enumerate(faiss_results):
        print(f"Result {i+1}:\n{doc.page_content}\n")

    print(f"\n--- Testing BM25 (Keyword) Retriever with query: '{test_query}' ---")
    bm25_results = bm25_retriever.invoke(test_query)
    for i, doc in enumerate(bm25_results):
        print(f"Result {i+1}:\n{doc.page_content}\n")
else:
    print("\nSkipping debugging session because knowledge base build failed.")



# CELL 7: INTERACTIVE QUESTION & ANSWERING (SIMPLIFIED)
# This cell uses a simplified but more robust retriever setup for the final Q&A.
if build_success:
    try:
        # We will use the Ensemble Retriever as it's the most robust
        print("\n--- Setting up Final RAG Chain with Ensemble Retriever ---")
        embeddings = HuggingFaceEmbeddings(model_name=EMBEDDING_MODEL_NAME)
        vector_store = FAISS.load_local(FAISS_INDEX_PATH, embeddings, allow_dangerous_deserialization=True)
        faiss_retriever = vector_store.as_retriever(search_kwargs={"k": 5})
        bm25_retriever = BM25Retriever.from_documents(text_chunks)
        bm25_retriever.k = 5

        ensemble_retriever = EnsembleRetriever(
            retrievers=[bm25_retriever, faiss_retriever],
            weights=[0.5, 0.5]
        )

        rag_chain = get_rag_pipeline(ensemble_retriever)

        print("\n--- Starting Interactive Q&A Session ---")
        print("Type your question and press Enter. Type 'exit' to end.")

        while True:
            query = input("\nYour Question: ")
            if query.lower() == 'exit':
                print("Exiting Q&A session.")
                break

            if query:
                print("⏳ Thinking...")
                try:
                    response = rag_chain.invoke(query)
                    print(f"\nAnswer: {response}")
                except Exception as e:
                    print(f"An error occurred: {e}")
            else:
                print("Please enter a question.")

    except (FileNotFoundError, ValueError) as e:
        print(f"\n❌ ERROR: {e}")
    except Exception as e:
        print(f"\n❌ An unexpected error occurred: {e}")
else:
    print("\nSkipping interactive session because knowledge base build failed.")
```
